```python
from __future__ import print_function
import urllib3 #  pip install -i https://pypi.anaconda.org/pypi/simple urllib3
import unicodedata
import pprint
from bs4 import BeautifulSoup, SoupStrainer, Comment
import warnings
import csv
warnings.filterwarnings('ignore')

#function to get a URL status
def checkURLStatus(url):
    http = urllib3.PoolManager()
    response = http.request('GET',url)
    return response

response=checkURLStatus("https://boston.craigslist.org/search/aap")
print(response.status)
```

The parent link gives the response status as 200 which means OK, therefore the link is active.

```python
soup=BeautifulSoup(response.data, "lxml")
allAnchor=soup.find_all('a')
print(allAnchor)
```

All anchor tags on the page are displayed above, now we will have to scrape the desired output from it.

```python
counter=0  #to count no. of records
data=[]
f = csv.writer(open('ScraperData.csv',"w", encoding="utf-8"),quotechar=" ")
f.writerow(["Link","Associated Text"])#insert header of csv file
for link in allAnchor:
    if(str(link['href']).startswith('http')):
        if(checkURLStatus(link['href']).status == 200):
            data.insert(0,str(link['href'])+'\t,"{}"'.format(str(link.string)))
            f.writerow([data[0]])
            counter+=1
            #print(data[0]) --uncomment to show each link in output
print("Done, exported "+str(counter)+" no. of links")
#It will take approx. 5 mins to run.
```

The code will export the data in a csv file with the name _"ScraperData.csv"_.
